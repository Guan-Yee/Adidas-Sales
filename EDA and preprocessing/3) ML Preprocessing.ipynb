##1) Read the dataset

```python
sales = pd.read_csv('/content/drive/MyDrive/Project/US Adidas Sales/Dataset/adidas_sales_new.csv')
```

```python
sales.info()
```

```python
sales.head(n = 5)
```

Drop the invoice Date

```python
sales = sales.drop(labels="Invoice Date", axis=1)
```

Drop the Retailer ID

```python
sales = sales.drop(labels="Retailer ID", axis=1)
```

```python
sales.info()
```

2020 has a lot of missing values due to covid 19
We will focus on 2021 instead

```python
sales_2021 = sales[sales["Year"] == 2021]
```

```python
sales_2021.info()
```

How many unique state and city are there?

```python
unique_states = sales_2021['State'].unique()
unique_cities = sales_2021['City'].unique()
```

```python
print(unique_states)
print(unique_cities)
```

```python
print(len(unique_states))
```

```python
print(len(unique_cities))
```

Create dummy variable for Retailer

```python
sales_2021 = pd.get_dummies(sales_2021, columns=['Retailer'])
```

```python
sales_2021.info()
```

Create Dummy variable for Region

```python
sales_2021 = pd.get_dummies(sales_2021, columns=['Region'])
```

```python
sales_2021.info()
```

Create dummy variable for Sales Method

```python
sales_2021 = pd.get_dummies(sales_2021, columns=['Sales Method'])
```

```python
sales_2021.info()
```

Create Dummy variable for the Product

```python
sales_2021 = pd.get_dummies(sales_2021, columns=['Product'])
```

```python
sales_2021.info()
```

Label Encode the Gender

```python
sales_2021['Gender'] = sales_2021['Gender'].replace({'Men': 0, 'Women': 1})
```

```python
sales_2021.head(n = 5)
```

Get the dummy variables for the State

```python
sales_2021 = pd.get_dummies(sales_2021, columns=['State'])
```

```python
sales_2021.info()
```

Drop the city column since the state is the low dimensional embedding

```python
sales_2021.drop('City', axis=1, inplace=True)
```

```python
sales_2021.info()
```

Drop the total sales and Operating Profit columns

```python
sales_2021.drop(['Total Sales','Operating Profit'], axis=1, inplace=True)
```

```python
sales_2021.info()
```

Assign outcome variables as units sales, and Operating Margin

```python
y1 = sales_2021["Units Sold"]
y2 = sales_2021["Operating Margin"]
```

Drop the columns respectively and create sales_2021_1 and sales_2021_2

```python
sales_2021_1 = sales_2021.drop(['Units Sold', 'Operating Margin','Year'], axis=1)
```

```python
sales_2021_2 = sales_2021.drop(['Operating Margin','Year'], axis=1)
```

Split the datasets into training and testing sets.

```python
from sklearn.model_selection import train_test_split
X_train1, X_test1, y_train1, y_test1 = train_test_split(sales_2021_1, 
                                                        y1, 
                                                        test_size=0.3, 
                                                        random_state=42)
```

```python
X_train1['Month'].value_counts()
```

```python
X_train2, X_test2, y_train2, y_test2 = train_test_split(sales_2021_2, 
                                                        y2, 
                                                        test_size=0.3, 
                                                        random_state=42)
```

```python
X_train2['Month'].value_counts()
```
